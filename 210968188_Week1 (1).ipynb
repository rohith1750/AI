Implement the CartPole environment for a certain number of steps

```python
import gym
from matplotlib import pyplot as plt

# Create the CartPole environment
env = gym.make('CartPole-v0', render_mode="human")

env.reset()  # Reset the environment to its initial state
num_steps = 100  # Define the total number of steps to run
cumulative_rewards_steps = []  # List to store cumulative rewards for each step
total_reward_steps = 0  # Variable to accumulate rewards for each step
steps=[]
for step in range(num_steps):
    env.render()  # Render the environment for visualization
    action = env.action_space.sample()  # Select a random action
    state, reward, done, info,_ = env.step(action)  # Apply the action
    total_reward_steps += reward  # Add reward to total
    cumulative_rewards_steps.append(total_reward_steps) 
    steps.append(step)
    # Append cumulative reward
    if done:
        env.reset()  # Reset the environment if pole falls

# Print the total reward accumulated over the steps
print(f"Total reward for fixed number of steps ({num_steps} steps): {total_reward_steps}")
env.close()  # Close the environment


```

```python
from matplotlib import pyplot as plt

plt.plot(steps, cumulative_rewards_steps)
plt.xlabel('Steps')
plt.ylabel('Cumulative Rewards')
plt.title('Cumulative Rewards over Steps')
plt.show()
```

Implement the CartPole environment for a certain number of episodes

```python
import gym
from matplotlib import pyplot as plt
# Create the CartPole environment
env = gym.make('CartPole-v0', render_mode="human")
```

```python
num_episodes = 20  # Define the total number of episodes
cumulative_rewards_episodes = []
total_reward_episodes = 0

# Loop through the defined number of episodes
for episode in range(num_episodes):
    state = env.reset()  # Reset environment for each new episode
    episode_reward = 0

    done = False
    while not done:
        env.render()
        action = env.action_space.sample()
        state, reward, done, info,_ = env.step(action)
        episode_reward += reward  # Add reward to episode total

    total_reward_episodes += episode_reward  # Add episode reward to total
    cumulative_rewards_episodes.append(total_reward_episodes)
    print(f"Episode {episode + 1} finished with reward: {episode_reward}")

print(f"Total reward for fixed number of episodes ({num_episodes} episodes): {total_reward_episodes}")

env.close()  # Close the environment

```

```python

plt.figure(figsize=(12, 6))

# Plot the cumulative rewards for each step
plt.subplot(1, 2, 1)
plt.plot(cumulative_rewards_steps, label='Cumulative Reward (Steps)')
plt.title('Cumulative Reward per Step')
plt.xlabel('Step')
plt.ylabel('Cumulative Reward')
plt.legend()

# Plot the cumulative rewards for each episode
plt.subplot(1, 2, 2)
plt.plot(cumulative_rewards_episodes, label='Cumulative Reward (Episodes)')
plt.title('Cumulative Reward per Episode')
plt.xlabel('Episode')
plt.ylabel('Cumulative Reward')
plt.legend()

plt.tight_layout()
plt.show()
```
